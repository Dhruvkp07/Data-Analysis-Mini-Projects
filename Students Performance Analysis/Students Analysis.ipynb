```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/StudentsPerformance (1).csv')
```

```python
df['Total']=df['math score']+df['reading score']+df['writing score']
```

```python
df['test preparation course'].unique()
```

```python
df[['math score', 'reading score', 'writing score']].describe()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

if numeric_df.shape[1] >= 3:
    plt.figure(figsize=(10, 8))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.tight_layout()
    plt.show()
else:
    print('Less than 3 numeric columns; skipping correlation heatmap.')
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df['gender'],df['Total'],alpha=0.4,color='green',width=.7)
plt.title('Male vs Female')
plt.xlabel('Gender')
plt.ylabel('Total')
plt.show()
```

```python

plt.figure(figsize=(10, 8))
sns.boxplot(x='race/ethnicity', y='Total', data=df,palette='pastel')
plt.title('Total Score Distribution by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Total Score')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='parental level of education', y='Total', data=df,palette='pastel')
```

```python
plt.figure(figsize=(10, 8))
plt.bar(df['test preparation course'],df['Total'],alpha=0.4,width=.7)


```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='test preparation course', y='math score', data=df,palette='pastel')
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='lunch', y='Total', data=df)
plt.title("Lunch Type vs Total Score")
plt.show()
```
